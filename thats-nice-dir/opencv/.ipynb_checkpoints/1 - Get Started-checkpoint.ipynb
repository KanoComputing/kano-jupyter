# Getting Started with OpenCV

Computers can look at pictures...and think about them just like humans.

This is called **computer vision**. 

We can use some cutting edge computer vision techniques with an Open Source library called **OpenCV**. 

Let's load the module!

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

## Loading an Image

Now let's load an image so that we can get the computer to think about it. Here's the picture we are going to use.

![](alex.jpg)

This is a picture of Kano's CEO and co-founder Alex Klein.

Let's get the computer to read the file into its memory:

```python
img = cv2.imread('alex.jpg',0)
```

## Edge Detection

Now that we have the image, let's get the computer to recognize the edges.

```python
edges = cv2.Canny(img,100,200)

plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

That sort of edge detection is very similar to the stuff that goes on in the Visual Cortex of the brain!
